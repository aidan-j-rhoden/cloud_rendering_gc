```python
#using google drive
from google.colab import drive
drive.mount("/gdrive")

print("")

#input blender file directory and install
get = str(input("Enter 'download' or 'install' from directories :- "))
if get == 'download':
  download = str(input("Enter Blender Download path link :- "))
  !wget $download
elif get == 'install':
  installer = str(input("Enter directory of blender :- "))
  !tar xf $installer
version = str(input('Enter blender version :- ')) 
print("")

#misc
!apt install libboost-all-dev
!apt install libgl1-mesa-dev

print("")

#selecting the render file 
filename = str(input("Enter directory of blend file to be rendered :- "))

print("")

#input render info
device = str(input("Enter render device :- "))
frame = str(input("Enter 's' for single frame or 'a' for animation :- "))

print("")

#cycles render device
if frame == "a":
  start = int(input("Enter starting frame no. :- "))
  end = int(input("Enter ending frame no. :- "))
  if device == "optix" or device == "OPTIX": 
    !sudo ./$version/blender -b $filename -noaudio  -E 'CYCLES' -o '//'Output'/Frame_#####' -s $start -e $end -a -- --cycles-device OPTIX
  elif device == "cuda" or device == "CUDA":
    cuda = str(input("Locate GPU.py file :- "))
    !sudo ./$version/blender -b $filename -noaudio -P $cuda  -E 'CYCLES' -o '//'Output'/Frame_#####' -s $start -e $end -a 
  else:
    print("Wrong input!")

elif frame == "s": 
  single = int(input("Enter frame no. :- "))
  if device == "optix" or device == "OPTIX": 
    !sudo ./$version/blender -b $filename -noaudio  -E 'CYCLES' -o '//'Output'/Frame_#####' -f $single -- --cycles-device OPTIX
  elif device == "cuda" or device == "CUDA":
    cuda = str(input("Locate GPU.py file :- "))
    !sudo ./$version/blender -b $filename -noaudio -P $cuda  -E 'CYCLES' -o '//'Output'/Frame_#####' -f $single 
  else:
    print("Wrong input!")
else:
  print("Wrong input!")
```
